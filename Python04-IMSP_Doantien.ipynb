```python
#1
products = {}

def add(prod,price):
    products[prod]=price

def change_price(prod,new_price):
    if prod in products.keys():
        products[prod]=new_price
    else:
        print("\nThis product does not existes\n")
def consult_price(prod):
    if prod in products.keys():
        print("\nThe price of the product {} is {}FCFA.\n".format(prod,products[prod]))
    else:
        print("\nThis product does not existes\n")
def get_prod_price():
    prod=input("\nThe name of the product :")
    price=int(input("Its price :"))
    return prod,price

def get_prod():
    prod=input("\nThe name of the product :")
    return prod
    
def menu():
    print("\n\n1-Add product\n2-Change price\n3-Consult price\4-End")
    choice=int(input("Your choice :"))
    return choice

end=False
while not end :
    choice = menu()
    if choice == 1:
        prod,price=get_prod_price()
        add(prod,price)
        print("Success add.")
    elif choice == 2:
        prod,price =get_prod_price()
        change_price(prod,price)
    elif choice == 3:
        prod=get_prod()
        consult_price(prod)
    elif choice == 4:
        print("Bye !")
        end=True
    else:
        print("Bad choice")

    
```

```python
#2
telephones = {}

def add(name,phone):
    telephones[name]=phone

def search_phone(name):
    if name in telephones.keys():
         print("\nThe telephone of   {} is {}.\n".format(name,telephones[name]))
        
    else:
        print("\nThis name does not existes\n")

def get_key(value):
    for k in telephones.keys():
        if telephones[k]==value:
            return k
    return None
        
def search_name(phone):
    k=get_key(phone)
    if k :
         print("\nThe owner of the number {} is {}.\n".format(phone,k))
        
    else:
        print("\nThis phone does not existes\n")
    
def print_phones():
    print("Liste of telephones \n_____________________")
    for n,p in  telephones.items():
        print("{} ==>> {}".format(n,p))
        
def get_name_phone():
    name=input("\nThe name :")
    phone=int(input("Its phone :"))
    return name,phone

def get_name():
    name=input("\nThe name:")
    return name

def get_phone():
    phone=int(input("\nThe phone:"))
    return phone
    
def menu():
    print("\n\n1-Add phone\n2-Search phone\n3-Search name\n4-List\n5-End")
    choice=int(input("Your choice :"))
    return choice

end=False
while not end :
    choice = menu()
    if choice == 1:
        name,phone=get_name_phone()
        add(name,phone)
        print("Success add.")
    elif choice == 2:
        name =get_name()
        search_phone(name)
    elif choice == 3:
        phone=get_phone()
        search_name(phone)
    elif choice == 4:
        print_phones()
    elif choice == 5:
        print("Bye !")
        end=True
    else:
        print("Bad choice")

    
```

```python
# 3
def integer(words):
    return [ len(w) for w in words]
    
words=["Toto","danser", "salut","representation","kao"]
number=integer(words)
print(number)
```

```python
# 4

def find_longest_word(words):
    number=integer(words)
    mx=number[0]
    i=0
    for j in range(len(number)):
        if number[j] > mx :
            mx=number[j]
            i=j
            
    return words[i]
print(find_longest_word(words))  
    
```

```python
# 5
dico={"merry":"god", "christmas":"jul", "and":"och", "happy":"gott", "new":"nytt", "year":"ar"}

def transate(words):
    return [dico[w.lower()] for w in words if w.lower() in dico.keys()]+ ["Don't know "+w for w in words if w.lower() not in dico.keys() ]

print(transate(["And","new","dflkkj"]))
```

```python
# 6
def is_member(x,lst):
    for i in lst:
        if i==x:
            return True
    return False

print(is_member(4,["j","5",45,42,[4,4]]))
```

```python
# 7
def histogram(lst):
    for i in lst:
        h=""
        for j in range(i):
            h=h+"*"
        print(h)
        
histogram([4,9,5])
```

```python
# 8

Amour = {
    "name": "Amour",
    "homework": [90.0,97.0,75.0,92.0],
    "quizzes": [88.0,40.0,94.0],
    "tests": [75.0,90.0]
}
Paix = {
    "name": "Paix",
    "homework": [100.0, 92.0, 98.0, 100.0],
    "quizzes": [82.0, 83.0, 91.0],
    "tests": [89.0, 97.0]
}
Foi = {
    "name": "Foi",
    "homework": [0.0, 87.0, 75.0, 22.0],
    "quizzes": [0.0, 75.0, 78.0],
    "tests": [100.0, 100.0]
}

students =[ Amour, Paix, Foi]

# print
print("Students infos\n__________________")
for s in students:
    for k,v in s.items():
        print("{} : {}".format(k,v))
    print("___________________________\n")

def average(lst):
    result = sum(lst)/len(lst)
    return result

def get_average(student):
    homework=average(student["homework"])
    quizzes=average(student["quizzes"])
    tests=average(student["tests"])
    
    avg =(10/100)*homework + (30/100)*(quizzes) + (60/100)*tests
    
    return avg


print("Amour average : ",get_average(Amour))

def get_letter_grade(score):
    if score >=90:
        return "A"
    elif score >=80:
        return "B"
    elif score >=70:
        return "C"
    elif score >=60:
        return "D"
    else:
        return "F"
print("His rade : ",get_letter_grade(get_average(Amour)))



def get_class_average(students):
    results=[]
    for student in students :
        results.append(get_average(student))
    return average(results)
print("Class average :", get_class_average(students))
print("Class grade :", get_letter_grade(get_class_average(students)))
```
